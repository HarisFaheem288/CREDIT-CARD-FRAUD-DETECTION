```python
import pandas as pd
df=pd.read_csv("fraud.csv")
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'first', 'last', 'street', 'trans_date_trans_time', 'dob', 'trans_num'])
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
df = pd.get_dummies(df, columns=['city','merchant', 'category', 'gender', 'job','state'], drop_first=True)
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('is_fraud', axis=1)  
y = df.is_fraud    
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score 
y_pred = model.predict(x_test)
print("Logistic Regression Accuracy is:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(random_state=42)
model1.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred1))
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=2, random_state=42)
model2.fit(x_train, y_train)
```

```python
y_pred2 = model2.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred2))
```
